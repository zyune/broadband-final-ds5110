```python
import pandas as pd
import requests
key='926b4f78d837ded44eeaaa6e4f00c0821952f74f'
```

```python
url='https://api.census.gov/data/timeseries/poverty/saipe?get=SAEPOVALL_PT,SAEMHI_PT,NAME&for=county:*&in=state:23&time=2010&key='+key
response=requests.get(url)
columns=response.json()[0]
data=response.json()[1:]
df1=pd.DataFrame(data,columns=columns)
df1
```

```python
import geopandas as gpd
gdf = gpd.read_file(
    'https://raw.githubusercontent.com/ds5010/broadband/main/src/county_boundaries/Maine_County_Boundary_Polygons_Dissolved_Feature.geojson')
# drop two useless columns that we are not gonna use 
gdf = gdf.drop(columns=['created_date'])
gdf = gdf.drop(columns=['last_edited_date'])
# add a new colum to geopandas dataframe, SAEMHI_PT stands for Median Household Income Estimate
gdf['SAEMHI_PT']=df1['SAEMHI_PT']
```

```python
gdf.explore(
     column="SAEMHI_PT", # make choropleth based on "BoroName" column
     tooltip="SAEMHI_PT", # show "BoroName" value in tooltip (on hover)
     popup=True, # show all values in popup (on click)
#      tiles="median income", # use "CartoDB positron" tiles
     cmap="Blues", # use "Set1" matplotlib colormap
     style_kwds=dict(color="grey") # use black outline
)
gdf.to_file('family_median_income.geojson')
```
